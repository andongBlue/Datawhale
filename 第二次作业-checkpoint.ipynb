## 数据的基本处理

**我们将‘status’的数据设为y_data，然后将其在表格中删除**

```python
y_data=data['status']
data.drop(['status'],axis=1,inplace=True)
```

**我看到缺值的数据，那么我通过pands将缺值的类型和数做一个统计，然后再做删除和填充的处理**

```python
dict_1=data.isnull().sum()
for key in list(dict_1.keys()):
    if dict_1[key]==0:
        del dict_1[key]
dict_1
```

**每一列的数据有4764个数据，如果缺失的数据量大于80%的话，就将这个数据删除**

```python
for key in list(dict_1.keys()):
    if dict_1[key]> int(4764*0.8):
        print("超过80%的数据是："+key)
        del dict_1[key]  
dict_1        
```

**没有超过80%的数据，那么保留所有数据，现在开始做缺失数据的填充**

**因为数据有个各种类型，将上面数据类型的结果和缺值的结果相比，我们发现缺失的值既有数值类型又有object类型，所以我们分开讨论**

**对于数值缺失，我们在这一列用均值填充，并打印object缺失的数据含义**

```python
for key in list(dict_1.keys()):
    if data[key].dtype!=object:
        data[key].fillna(data[key].mean(),inplace=True)
    else:print("缺失的object数据是什么含义："+key)

```

```python
data['reg_preference_for_trad'].value_counts()
```

**我们将reg_preference_for_trad数据缺额全部填充为其他城市**

```python
data['reg_preference_for_trad'].fillna("其他城市",inplace=True)
```

**对于latest_query_time和loans_latest_time我们都用众数填充**

```python
data['latest_query_time'].fillna(data['latest_query_time'].mode()[0],inplace=True)
data['loans_latest_time'].fillna(data['loans_latest_time'].mode()[0],inplace=True)
```

**检查是否填充完毕**

```python
data.isnull().sum()
```

## 数据标准化/归一化

```python
for i in data.columns:
    if data[i].dtype==object:
            print(i)
    else:
        data[i]=(data[i]-data[i].min())/(data[i].max()-data[i].min())
data.head() #显示头部的数据
```

## 进行特质编码 One-hot encoding

```python
data=data.join(pd.get_dummies(data.reg_preference_for_trad))
data.drop(data['reg_preference_for_trad'],axis=1,inplace=True)
```

**时间分为latest_query_time和loans_latest_time，我们将其格式化**

```python
data["咨询时间"]=pd.to_datetime(data['latest_query_time'],format='%Y-%m-%d')
data["贷款时间"]=pd.to_datetime(data['loans_latest_time'],format='%Y-%m-%d')
data["贷款时间"].tail()
```

```python
data['年1'] = [i.year for i in data["咨询时间"]]
data['月1'] = [i.month for i in data["咨询时间"]]

data['年2'] = [i.year for i in data["贷款时间"]]
data['月2'] = [i.month for i in data["贷款时间"]]
```

```python
cols = ['年1','月1','年2','月2']
for col in cols:
    data = data.join(pd.get_dummies(data[col]),how='left', lsuffix='_left', rsuffix='_right')
```

```python
cols = ['年1','月1','年2','月2','贷款时间','咨询时间','latest_query_time','loans_latest_time']
for col in cols:
    data.drop([col],axis = 1,inplace = True)
```

```python
data.drop(['source'],axis = 1,inplace = True)
data.drop(['trade_no'],axis = 1,inplace = True)
data
```
